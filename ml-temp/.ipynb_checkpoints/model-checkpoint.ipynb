```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
import joblib
sns.set()
```

```python
data = pd.read_csv('database_data.csv')
data = data.drop('Unnamed: 0', axis = 1)
data = data.set_index('id')
data
```

```python
data.isna().sum()
```

# Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
encoder = OrdinalEncoder()
xy = encoder.fit_transform(data['type_of_consultant'].values.reshape(-1,1))
xy
```

```python
he = OneHotEncoder()
ab = he.fit_transform(xy).toarray()
ab
```

```python
data[['Career','Doctor','Education','Lawyer']] = ab
data
```

# Selecting Features

```python
X = data.iloc[:,[4,5,6,12,13,14,15,16,17]]
y = pd.DataFrame([ 0 for i in range(2006)])
X
```

# Training Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model =KNeighborsClassifier()
model.fit(X,y)
model
```

```python
X_predict = X.iloc[1500:1501,:]
X_predict
```

```python
result = model.kneighbors(X_predict,n_neighbors=10)
result = np.array(result)
result[1]
```

```python
list(map(int,result[1].reshape(-1,1)))
```

```python
X.iloc[list(map(int,result[1].reshape(-1,1))),:]
```

```python
joblib.dump(model,'model_final.sav')
```
