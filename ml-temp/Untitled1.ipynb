```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
import joblib
sns.set()
```

```python
data = pd.read_csv('database_data.csv')
data = data.drop('Unnamed: 0', axis = 1)
data = data.set_index('id')
data
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
encoder = OrdinalEncoder()
encoder.fit([['Career'],['Doctor'],['Education'],['Lawyer']])
joblib.dump(encoder,'LabelEncoder.sav')
```

```python
xy = encoder.transform(data.iloc[4:5,:]['type_of_consultant'].values.reshape(-1,1))
print(xy)
he = OneHotEncoder()
ab = he.fit_transform(xy).toarray()
ab
```
